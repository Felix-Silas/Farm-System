```python
# 1번 문제
number = int(input('자연수 n을 입력하세요. --> '))
value = 1

for n in range(1, number + 1):
    value *= n
    
print('factorial(n) =',value)
```

```python
# 2번 문제
str = input('문자열을 입력하세요. --> ')

for i in range(len(str)):
    if str[i] == ',':
        print(str[i], end ='')
    elif i == len(str) - 1:
        print(str[i].upper())   
    else:
        print(str[i].upper(), end = ',')
```

```python
# 3번 문제
value = int(input('2진수로 변환할 양수를 입력하시오. --> '))

b = format(value, 'b')
print(b)
```

```python
# 4번 문제
str = input('정수 A와 B, 그리고 사칙연산을 입력해주세요 --> ')
first = int(str.split()[0])
second = int(str.split()[1])
opr = str.split()[2]

result = 0

if opr == '+':
    result = first + second
    print(result)
elif opr == '-':
    result = first - second
    print(result)
elif opr == '*':
    result = first * second 
    print(result)
elif opr == '/':
    result = first / second
    print(int(result))
```

```python
# 5번 문제
a=input('계산할 식을 입력하시오. ')
stack=[]
res=''
for x in a:
    if x.isdecimal():
        res+=x
    else:
        if x=='*' or x=='/':
            while stack and (stack[-1]=='*' or stack[-1]=='/'):
                res+=stack.pop()
            stack.append(x)
        elif x=='+' or x=='-':
            while stack and stack[-1]!='(':
                res+=stack.pop()
            stack.append(x)
            
while stack:
    res+=stack.pop()
    
b = res
stack=[]

for x in b:
    if x.isdecimal():
        stack.append(int(x))
    else:
        if x=='+':
            n1=stack.pop()
            n2=stack.pop()
            stack.append(n2+n1)
        elif x=='-':
            n1=stack.pop()
            n2=stack.pop()
            stack.append(n2-n1)
        elif x=='*':
            n1=stack.pop()
            n2=stack.pop()
            stack.append(n2*n1)
        elif x=='/':
            n1=stack.pop()
            n2=stack.pop()
            stack.append(n2/n1)
    
print(res, int(stack[0]))
```
